```python
url = 'https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/anagrafica-vaccini-summary-latest.csv'

df = pd.read_csv(url)
df
```

Visualizzare il numero di vaccinati nella fascia di 12-19

```python
df.loc[df['fascia_anagrafica'] == '12-19', 'totale']
```

Visualizzare il numero di vaccinati maschi e femmine

```python
df.loc[df['fascia_anagrafica'] == '12-19', ['sesso_maschile','sesso_femminile']]
```

fascie età dove ci sono più di 10 milioni di vaccinazioni


```python
df.loc[df['totale'] >= 10000000,'fascia_anagrafica']
```

fascie di età dove sono stati vaccinati maschi e femmine da 10-20 milioni

```python
df.loc[(df['totale'] < 2000000) & (df['totale'] > 1000000) 'fascia_anagrafica')]
```

Aggiungere una colonna '%sesso maschile' dove inserire la percentuale di maschi vaccinati sul totale

```python
df['%sesso maschile'] = (df['sesso_maschile']*100) / df['totale']
df
```

Aggiungere una colonna '%sesso maschile' dove inserire la percentuale di maschi vaccinati sul totale

```python
df['%sesso femminile'] = (df['sesso_femminile']*100) / df['totale']
df
```

Visualizzare le fascie di età in cui la percentuale di maschi vaccinati è maggiore della percentuale di femmine vaccinate

```python
df.loc[df['%sesso maschile'] > df['%sesso femminile']]
```

Visualizzare le fascie di età dove la dose aggiuntiva è minore della dose bookster

```python
df.loc[df['dose_aggiuntiva'] < df['dose_booster']]['fascia_anagrafica']
```
