scrivere una funzione chiamata prodotto che data una lista di numeri interi restituisca il prodotto di tutti i suoi elementi. Ad esempio se la lista fosse [2, 5, 3] la funzione restituirebbe 30


```python
def prodotto(l):
  totale = 1
  
  for e in l:
   totale = totale*e
  return totale

lista = [int(ele) for ele in input("inserire i numeri: ").split()]
prodotto(lista)
```

scrivere un programma che, utilizzando una lambda function e la funzione map, elevi alla terza tutti gli elementi della lista. La lista deve essere caricata dall’utente.


```python
lista = [int(ele) for ele in input("inserire i numeri: ").split()]
listafinale = list(map(lambda x: x**3, lista))
print(listafinale)
```
