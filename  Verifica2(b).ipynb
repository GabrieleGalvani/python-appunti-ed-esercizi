```python
#3.Es
parola = input()
vocali = ['a','e','i','o','u']
x = 0
for lettera in parola:
  if lettera not in vocali:
    print('parola non di sole vocali')
    break
  else:
    x+=1
  if x == len(parola):
    print(parola + ' contiene solo vocali')
```

```python
#4.Es
lst = input().split()
parole = []
val = int(input())
for par in lst:
  if len(par)<=val:
    parole.append(par)
print(parole)
```
