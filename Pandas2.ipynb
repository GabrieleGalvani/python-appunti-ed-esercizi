```python
import pandas as pd
```

```python
url = 'https://shanelynnwebsite-mid9n9g1q9y8tt.netdna-ssl.com/wp-content/uploads/2015/06/phone_data.csv'
df = pd.read_csv(url)
df
```

```python
df.groupby('month').first()
```

```python
df.groupby('month').groups.keys()
```

```python
df.groupby('month')['duration'].sum()
```

```python
df.groupby('month')['duration'].count()
```

```python
df.groupby('month')['duration'].sum()
```

```python
operations = {
    'duration': sum,
    'network_type':"count",
    'date':'first'
}

df.groupby(['month','item']).agg(operations)
```

```python
operations = {
    'duration': ['mean', 'sum'],
    'network_type':"count",
    'date':'first'
}

df.groupby(['month','item']).agg(operations)
```

```python
operations = {
    'duration': ['mean', 'sum'],
    'network_type':"count",
    'date':'nunique'
}

df.groupby(['month','item']).agg(operations)
```

```python
gb = df.groupby('month').apply(lambda x: (x['duration'] >= 1000))
gb
```
